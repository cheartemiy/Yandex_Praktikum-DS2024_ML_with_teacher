## ВВЕДЕНИЕ 

**Описание проекта:**

Интернет-магазин «В один клик» занимается продажей товаров различных категорий. Недавний отчет показал снижение активности покупателей. Привлекать новых клиентов стало менее эффективно, поэтому важно сосредоточиться на удержании активности постоянных клиентов через персонализированные предложения. В качестве стажера в отделе цифровых технологий, нам предстоит разработать решение для персонализации предложений, которое позволит персонализировать предложения постоянным клиентам, чтобы увеличить их покупательскую активность.

**Как решать задачу:**

1. **Промаркировать уровень финансовой активности постоянных клиентов**, определяя два основных уровня:
   - **Снизилась**: клиент стал покупать меньше товаров.
   - **Прежний уровень**: клиент продолжает покупать товары на прежнем уровне.
   
   
2. **Собрать данные по клиентам, разделив на группы:**
    - **Коммуникационные признаки**: Описывают взаимодействие с клиентом.
    - **Продуктовое поведение**: Типы покупаемых товаров и частота покупок.
    - **Покупательское поведение**: Сумма трат в магазине.
    - **Поведение на сайте**: Активность на сайте, включая просмотр страниц и время на сайте.


3. **Моделирование:** 
    - **Построение прогностической модели** для предсказания вероятности снижения покупательской активности на следующие три месяца.
    - **Анализ финансовой информации** о прибыльности клиента от финансового департамента, включая доходы от каждого покупателя за последние три месяца.
    - **Сегментация покупателей** на основе данных модели и финансовой информации, с целью разработки персонализированных предложений.

# **Описание данных:** 

- Данные разбиты на четыре таблицы

1. **market_file.csv**: Основные данные о поведении и взаимодействии покупателей с магазином.

## Исследовательский анализ данных (EDA)

```python
def numerical_stats(data): 
    # Функция принимает только количественные значения
    numerical_features = data.select_dtypes(include=['number']).columns
    
    for numerical_feature in numerical_features: 
        # Строим boxplot и гистограмму
        fig, ax = plt.subplots(1, 2, figsize=(20, 6))
        
        # Boxplot
        sns.boxplot(x=data[numerical_feature], ax=ax[0])
        ax[0].set_title(f'Boxplot - {numerical_feature}', size=20)
        ax[0].set_xlabel('Значение', size=14)
        
        # Гистограмма 
        ax[1].hist(data[numerical_feature], bins=30, edgecolor='k')
        ax[1].set_title(f'Распределение - {numerical_feature}', size=20)
        ax[1].set_xlabel('Значение', size=14)
        ax[1].set_ylabel('Частота', size=14)
        
        plt.show()
        
        # Describe
        stats = data[numerical_feature].describe()
        print(stats)
        
        # Вывод среднего и медианы
        mean = data[numerical_feature].mean()
        median = data[numerical_feature].median()
        print(f'{numerical_feature} - Среднее: {mean}, Медиана: {median}')
        print('_________________________________________________________________________________________________________________')
```

```python
# Функция для отображения категориальных значений с сортировкой
def categorial_stats(data):
    # Функция принимает только категориальные значения
    categorical_features = data.select_dtypes(exclude=['number']).columns
    
    for categorical_feature in categorical_features:
        # Получаем порядок категорий по количеству
        val_counts = data[categorical_feature].value_counts().index
        
        plt.figure(figsize=(14, 6))
        sns.countplot(y=categorical_feature, data=data, order=val_counts)
        plt.title(f'Распределение по {categorical_feature}', size=14)
        plt.xlabel('Количество', size=14)
        plt.ylabel(categorical_feature, size=14)
        plt.show()
        
        print('_________________________________________________________________________________________________________________')
```

## Использование пайплайнов

Наличие дисбаланса классов в данных о покупательской активности может оказать влияние на процесс обучения и точность прогнозируемой модели. Однако, учитывая, что дисбаланс классов не является значительным, применение стратификации представляется разумным решением. Стратификация поможет обеспечить равномерное распределение классов целевого признака в обучающей и тестовой выборках, что способствует улучшению обобщающей способности модели и повышению качества прогнозов.

```python
# Установка констант для воспроизводимости и размера тестовой выборки
RANDOM_STATE = 42
TEST_SIZE = 0.25

# Извлечение входных и целевых признаков
# Входные признаки - все столбцы, кроме целевого признака и идентификатора клиента
X = market_final.drop('покупательская_активность', axis=1)
# Целевой признак - покупательская активность
y = market_final['покупательская_активность']

# Разделение данных на обучающую и тестовую выборки с учетом стратификации
X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=TEST_SIZE, 
    random_state=RANDOM_STATE, 
    stratify=y
) 

# Визуализация распределения целевого признака для исходных данных и для выборок
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
axes = axes.flatten()

# График для исходного распределения целевого признака
y.value_counts().plot(kind='bar', ax=axes[0], color='skyblue')
axes[0].set_title("Распределение в исходных данных")
axes[0].set_ylabel('Количество наблюдений')

# График для распределения целевого признака в обучающей выборке
y_train.value_counts().plot(kind='bar', ax=axes[1], color='lightgreen')
axes[1].set_title("Распределение в обучающей выборке")

# График для распределения целевого признака в тестовой выборке
y_test.value_counts().plot(kind='bar', ax=axes[2], color='salmon')
axes[2].set_title("Распределение в тестовой выборке")

# Устанавливаем одинаковые масштабы осей для всех графиков
max_y = max(y.value_counts().max(), y_train.value_counts().max(), y_test.value_counts().max())
for ax in axes:
    ax.set_ylim(0, max_y)

plt.show()  
```

### Построение пайплайна для оптимизации модели

Для выбора оптимальной модели и настройки гиперпараметров будет создан пайплайн. В его рамках будут тестироваться следующие модели и их настройки:

- **DecisionTreeClassifier**: 
  - `max_depth` в диапазоне от 2 до 6,
  - `max_features` в диапазоне от 2 до 6.

- **KNeighborsClassifier**:
  - `n_neighbors` от 2 до 6.

- **LogisticRegression**:
  - `solver='liblinear'`,
  - лассо-регуляризация с `C` от 1 до 6.

- **SVC**:
  - `kernel=['linear', 'poly, 'rbf', 'sigmoid']`,
  - перебор параметра `C` от 1 до 6.

После определения лучшей модели и её параметров, будет измерена метрика ROC-AUC на обучающей и тестовой выборках. Использование ROC-AUC обусловлено её применимостью для оценки моделей бинарной классификации при наличии дисбаланса классов.


```python
# создаём списки с названиями признаков
ohe_columns = ['популярная_категория']
ord_columns = ['тип_сервиса', 'разрешить_сообщать']
num_columns = ['маркет_актив_6_мес', 
              'маркет_актив_тек_мес',
              'длительность',
              'акционные_покупки',
              'средний_просмотр_категорий_за_визит',
              'неоплаченные_продукты_штук_квартал',
              'ошибка_сервиса',
              'страниц_за_визит',
              'выручка_предыдущий_месяц',
              'выручка_препредыдущий_месяц',
              'выручка_текущий_месяц',
              'время_предыдущий_месяц',
              'время_текущий_месяц']

ohe_pipe = Pipeline(
    [('ohe', OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False))]
    )

ord_pipe = Pipeline(
    [('ord',  OrdinalEncoder(
                categories=[
                    ['стандарт', 'премиум'], 
                    ['да', 'нет']
                ],
        handle_unknown='use_encoded_value', unknown_value=np.nan))
    ]
)

# создаём общий пайплайн для подготовки данных
data_preprocessor = ColumnTransformer(
    [('ohe', ohe_pipe, ohe_columns),
     ('ord', ord_pipe, ord_columns),
     ('num', MinMaxScaler(), num_columns)
    ], 
    remainder='passthrough'
)


# создаём итоговый пайплайн: подготовка данных и модель
pipe_final = Pipeline([
    ('preprocessor', data_preprocessor),
    ('models', DecisionTreeClassifier(random_state=RANDOM_STATE))
])

param_grid = [
    # словарь для модели KNeighborsClassifier() 
    {
        # название модели
        'models': [KNeighborsClassifier()],
        'models__n_neighbors': range(2, 6),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), RobustScaler(), 'passthrough']   
    },
    # словарь для модели DecisionTreeClassifier()
    {
        'models': [DecisionTreeClassifier(random_state=RANDOM_STATE)],
        'models__max_depth': range(2, 6),
        'models__max_features': range(2, 6),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), RobustScaler(), 'passthrough']  
    },
    # словарь для модели SVC()
    {
        'models': [SVC(random_state=RANDOM_STATE, probability=True)],
        'models__kernel': ['linear', 'poly', 'rbf', 'sigmoid'],
        'models__C': range(1, 6),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), RobustScaler(), 'passthrough']  

    },
    # словарь для модели LogisticRegression()
    {
        'models': [LogisticRegression(random_state=RANDOM_STATE, 
                                     solver='liblinear',
                                     penalty='l1'
                                     )],
                'models__C': range(1, 6),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), RobustScaler(), 'passthrough']  
    }
]

randomized_search = RandomizedSearchCV(
    pipe_final, 
    param_grid, 
    cv=5,
    scoring='roc_auc',
    n_iter=5,
    random_state=RANDOM_STATE,
    n_jobs=-1
)
randomized_search.fit(X_train, y_train)
print('Лучшая модель и её параметры:\n\n', randomized_search.best_estimator_)
print ('Метрика лучшей модели на тренировочной выборке:', randomized_search.best_score_)

# проверим работу модели на тестовой выборке
X_test_pred_proba = randomized_search.predict_proba(X_test)[:, 1]
print(f'Метрика ROC-AUC на тестовой выборке: {roc_auc_score(y_test, X_test_pred_proba)}')
```

```python
#Сравнение результатов
pd.DataFrame(randomized_search.cv_results_)[
    ['std_test_score', 'rank_test_score', 'param_models', 'mean_test_score','params']
].sort_values('rank_test_score')
```

### Лучшая модель

В процессе оптимизации моделей для задачи бинарной классификации с учетом дисбаланса классов, модель **SVC** с параметром `C=5` продемонстрировала наилучшие результаты. Метрика ROC-AUC на тренировочной выборке составила ~0.907, а на тестовой — 0.849, что свидетельствует о высокой обобщающей способности модели.

Выбор был сделан в пользу линейного ядра (`kernel='linear'`) по сравнению с RBF, при одинаковом значении `C=5`, по следующим причинам:

- **Простота и интерпретируемость**: Линейное ядро является более простым для понимания и интерпретации, что особенно важно при необходимости объяснения модели заинтересованным сторонам.
- **Вычислительная эффективность**: Линейное ядро обычно требует меньше вычислительных ресурсов по сравнению с RBF, особенно в условиях большого количества признаков.

Таким образом, предпочтение линейному ядру обусловлено желанием достичь оптимального соотношения между производительностью модели и ее вычислительной эффективностью, а также обеспечить интерпретируемость результатов.


```python
#OneHotEncoder()
ohe_encoder = OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False)
X_train_ohe = ohe_encoder.fit_transform(X_train[ohe_columns])
X_test_ohe = ohe_encoder.transform(X_test[ohe_columns])
ohe_encoder_col_names = ohe_encoder.get_feature_names_out()

#OrdinalEncoder()
ord_encoder = OrdinalEncoder(categories=[
                    ['стандарт', 'премиум'], 
                    ['да', 'нет']
                ],
        handle_unknown='use_encoded_value', unknown_value=np.nan)
X_train_ord = ord_encoder.fit_transform(X_train[ord_columns])
X_test_ord = ord_encoder.transform(X_test[ord_columns])
ord_encoder_col_names = ord_encoder.get_feature_names_out()


# масштабируем количественные признаки с помощью StandardScaler()
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train[num_columns])
X_test_scaled = scaler.transform(X_test[num_columns])

# создаём тренировочный и тестовый датафреймы из закодированных и отмасштабированных данных

X_train_ord = pd.DataFrame(X_train_ord, columns=ord_encoder_col_names)
X_test_ord = pd.DataFrame(X_test_ord, columns=ord_encoder_col_names)\

X_train_ohe = pd.DataFrame(X_train_ohe, columns=ohe_encoder_col_names)
X_test_ohe = pd.DataFrame(X_test_ohe, columns=ohe_encoder_col_names)

X_train_scaled = pd.DataFrame(X_train_scaled, columns=num_columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=num_columns)

X_train_prepared = pd.concat([X_train_ohe, X_train_ord, X_train_scaled], axis=1)
X_test_prepared = pd.concat([X_test_ohe, X_test_ord, X_test_scaled], axis=1)

print(y_train.shape, y_test.shape)
print(X_train_prepared.shape, X_test_prepared.shape)
```

```python
# Объявляем лучшую модель SVC с силой 5
clf = SVC(C=5, kernel='linear', probability=True, random_state=42)

# Обучаем модель на тренировочной выборке
clf.fit(X_train_prepared, y_train)

# Предсказываем и оцениваем модель на тестовой выборке
y_pred = clf.predict(X_test_prepared)
y_probas = clf.predict_proba(X_test_prepared)[:, 1]

# Считаем метрику ROC-AUC для тренировочной выборки с помощью кросс-валидации
roc_auc_cv = cross_val_score(clf, X_train_prepared, y_train, scoring='roc_auc', cv=5).mean()

# Считаем метрику ROC-AUC и accuracy для тестовой выборки
roc_auc_test = roc_auc_score(y_test, y_probas)
accuracy_test = accuracy_score(y_test, y_pred)
f1_score_test = f1_score(y_test, y_pred)

# Выводим результаты
print(f'ROC-AUC на тренировочной выборке (CV): {roc_auc_cv:.4f}')
print(f'ROC-AUC на тестовой выборке: {roc_auc_test:.4f}')
print(f'Accuracy на тестовой выборке: {accuracy_test:.4f}')
print(f'F1-score на тестовой выборке: {f1_score_test:.4f}')
```

## Сегментация покупателей

**Вероятность снижения покупательской активности (используя `probability`):**
- **Низкая вероятность снижения активности**: если `probability` > 0.5, значит предсказание модели положительное, т.е. активность не снизится.
- **Высокая вероятность снижения активности**: если `probability` <= 0.5, значит предсказание модели отрицательное, т.е. активность снизится.

**Уровень прибыли:**
- **Высокая прибыль**: если значение `прибыль` выше медианы по всему набору данных.
- **Низкая прибыль**: если значение `прибыль` ниже или равно медиане по всему набору данных.

**Шаги анализа:**
1. **Определение медианных значений**: Определим медианное значение `прибыль` для разделения на высокую и низкую прибыль.

3. **Подсчет количества покупателей в сегментах**: Подсчитаем количество покупателей в каждом сегменте.
